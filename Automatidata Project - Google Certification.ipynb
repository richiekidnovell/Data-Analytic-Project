A **box plot (or box-and-whisker plot)** shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset, while the whiskers extend to show the rest of the distribution, except for points that are determined to be outliers.

## EXPERIMENT

To confirm conclusion, consider the following experiment:
1. Create a sample of coordinates from a normal distribution&mdash;in this case 1,500 pairs of points from a normal distribution with a mean of 10 and a standard deviation of 5
2. Calculate the distance between each pair of coordinates 
3. Group the coordinates by endpoint and calculate the mean distance between that endpoint and all other points it was paired with
4. Plot the mean distance for each unique endpoint

```python
# Set the seed for reproducibility
np.random.seed(0)

# Generate 1500 pairs of points from a normal distribution
num_pairs = 1500
points = np.random.normal(loc=10, scale=5, size=(num_pairs, 2, 2))

# Calculate the Euclidean distance between each pair of points
distances = np.linalg.norm(points[:, 0, :] - points[:, 1, :], axis=1)

# Create a DataFrame with the endpoints and distances
data = pd.DataFrame({
    'endpoint_1': points[:, 0, :].tolist(),
    'endpoint_2': points[:, 1, :].tolist(),
    'distance': distances
})

# Display the first few rows of the DataFrame
data.head()

```

Next, we'll "flatten" this DataFrame so that each row represents an endpoint (either 'endpoint_1' or 'endpoint_2') and its associated distance. This will allow us to group the data by endpoint and calculate the mean distance for each unique endpoint.

```python
# Flatten the DataFrame
df_flat = pd.concat([
    data[['endpoint_1', 'distance']].rename(columns={'endpoint_1': 'endpoint'}),
    data[['endpoint_2', 'distance']].rename(columns={'endpoint_2': 'endpoint'})
])

```

```python
df_flat['endpoint'] = df_flat['endpoint'].apply(tuple)

# Group by endpoint and calculate mean distance
mean_distance_by_endpoint = df_flat.groupby('endpoint')['distance'].mean()

# Display the mean distance for each unique endpoint
mean_distance_by_endpoint
```

Now we have successfully calculated the mean distance for each unique endpoint. Each row of the output represents an endpoint (a 2D coordinate) and the mean distance between that endpoint and all other points it was connected with.

Lastly, let's plot the mean distance for each unique endpoint. To make the plot more readable, we'll only plot a subset of the endpoints. Let's say, the 50 endpoints with the highest mean distance.

```python
# Get the 50 endpoints with the highest mean distance
top_endpoints = mean_distance_by_endpoint.nlargest(50)

# Create a bar plot for the mean distance for each unique endpoint
plt.figure(figsize=(10, 10))
plt.barh(range(len(top_endpoints)), top_endpoints.values, color='skyblue', edgecolor='black')
plt.yticks(range(len(top_endpoints)), top_endpoints.index)

# Set the title and labels
plt.title('Mean Distance for Each Unique Endpoint')
plt.xlabel('Mean Distance')
plt.ylabel('Endpoint')

# Show the plot
plt.show()
```
